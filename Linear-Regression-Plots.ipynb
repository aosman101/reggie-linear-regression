# Reggie-Linear-Regression Plots

This notebook includes five distinct plot types using NumPy and Matplotlib. Each plot is created in its own cell and saved under a `figures/` directory as both PNG and SVG for easy inclusion in a repository.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Reproducibility & output dir
np.random.seed(42)
Path("figures").mkdir(exist_ok = True)

def savefig(name):
    for ext in ("png", "svg"):
        plt.savefig(Path("figures") / f"{name}.{ext}", bbox_inches = "tight")
```

```python
# 1) Line Plot: Noisy Sine Wave
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 200)
y = np.sin(t) + 0.3 * np.random.randn(t.size)

plt.figure()
plt.plot(t, y, label="Noisy sine")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Line Plot: Noisy Sine Wave")
plt.legend()
plt.tight_layout()
savefig("plot01_line")
plt.show()
```

### 1) Line Plot: Noisy Sine Wave

- This plot shows a noisy sine wave over time. It demonstrates how to create a basic line plot with labels, a title, and a legend. The `savefig` function is used to save the plot in both PNG and SVG formats.

```python
# 2) Scatter with Linear Fit
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = 0.8 * x + 2 + 0.8 * np.random.randn(x.size)
m, b = np.polyfit(x, y, 1)

plt.figure()
plt.scatter(x, y, alpha=0.7, label="Points")
plt.plot([x.min(), x.max()], [m * x.min() + b, m * x.max() + b], linewidth=2,
         label=f"Fit: y={m:.2f}x+{b:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter with Linear Fit")
plt.legend()
plt.tight_layout()
savefig("plot02_scatter_fit")
plt.show()
```

### 2) Scatter with Linear Fit

- This plot displays a scatter plot of points with a linear fit line. It illustrates how to generate random data, perform a linear regression using `np.polyfit`, and plot both the data points and the fitted line. The equation of the fitted line is also included in the legend.

```python
# 3) Grouped Bar Chart
import numpy as np
import matplotlib.pyplot as plt

labels = ["Q1", "Q2", "Q3", "Q4"]
x = np.arange(len(labels))
a = np.array([24, 30, 21, 27])
b = np.array([20, 28, 25, 31])
w = 0.35

plt.figure()
rects1 = plt.bar(x - w/2, a, width=w, label="Product A")
rects2 = plt.bar(x + w/2, b, width=w, label="Product B")
plt.xticks(x, labels)
plt.ylabel("Units")
plt.title("Grouped Bar Chart")
plt.legend()

# Annotate bars
for rects in (rects1, rects2):
    for r in rects:
        h = r.get_height()
        plt.text(r.get_x() + r.get_width()/2, h, f"{h}", ha="center", va="bottom", fontsize=8)

plt.tight_layout()
savefig("plot03_bar_grouped")
plt.show()
```

### 3) Grouped Bar Chart

- This plot presents a grouped bar chart comparing two products across different quarters. It shows how to create grouped bars using `plt.bar` with an offset and width, set x-axis labels, and annotate the bars with their respective values.

```python
# 4) Box Plot: Four Distributions
import numpy as np
import matplotlib.pyplot as plt

g1 = np.random.normal(loc=0.0, scale=1.0, size=300)
g2 = np.random.normal(loc=0.5, scale=1.5, size=300)
g3 = np.random.normal(loc=-0.5, scale=0.7, size=300)
g4 = np.random.normal(loc=1.0, scale=1.2, size=300)

plt.figure()
plt.boxplot([g1, g2, g3, g4], notch=True, showmeans=True)
plt.xticks([1, 2, 3, 4], ["A", "B", "C", "D"])
plt.ylabel("Value")
plt.title("Box Plot: Four Distributions")
plt.tight_layout()
savefig("plot04_boxplot")
plt.show()
```


### 4) Box Plot: Four Distributions

- This plot visualises the distribution of four different groups using box plots. It demonstrates how to create box plots with notches and mean markers, set x-axis labels for each group, and display the distribution of the data.


```python
# 5) Heatmap: Correlation Matrix
import numpy as np
import matplotlib.pyplot as plt

N = 300
z = np.random.normal(size=N)
x1 = z + 0.3 * np.random.normal(size=N)
x2 = -0.6 * z + 0.3 * np.random.normal(size=N)
x3 = 0.5 * x1 + 0.5 * np.random.normal(size=N)
x4 = np.random.normal(size=N)

X = np.column_stack([x1, x2, x3, x4])
labels = ["x1", "x2", "x3", "x4"]
corr = np.corrcoef(X, rowvar=False)

plt.figure()
im = plt.imshow(corr, vmin=-1, vmax=1)
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.xticks(np.arange(len(labels)), labels)
plt.yticks(np.arange(len(labels)), labels)
plt.title("Heatmap: Correlation Matrix")

# Annotate cells
for i in range(corr.shape[0]):
    for j in range(corr.shape[1]):
        plt.text(j, i, f"{corr[i, j]:.2f}", ha="center", va="center", fontsize=8)

plt.tight_layout()
savefig("plot05_heatmap_corr")
plt.show()
```

### 5) Heatmap: Correlation Matrix

 - This plot displays a heatmap of a correlation matrix. It shows how to calculate the correlation matrix of a dataset using `np.corrcoef`, visualise it using `plt.imshow` with a colourbar, and annotate the cells with the correlation values.
